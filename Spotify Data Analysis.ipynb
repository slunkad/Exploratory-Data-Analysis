### Converting duration in milliseconds to seconds

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x : round(x/1000))
df_tracks.drop("duration_ms" , inplace = True , axis=1)
```

```python
df_tracks["duration"]
#same as df_tracks.duration
```

Creating correlation matrix

```python
correlationmatrix = df_tracks.corr()
round(correlationmatrix,2 )
```

Creating a heatmap

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap= sns.heatmap(corr_df , annot= True, fmt= ".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

Regression PLot

```python
#Creating a sample of 0.4% of data
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
                             
```

Regression plot between loudness and energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness" , x="energy", color="c").set_title("Loudness v/s Energy Correlation")
```

```python
#very high positive correlation between loudness and energy
```

Regression plot between popularity and danceability

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity" , x="danceability", color="pink").set_title("Popularity v/s Danceability Correlation")
```

```python
#the regression line has positive slope, but the correlation is weak
```

Regression plot between popularity and acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity" , x="acousticness", color="green").set_title("Popularity v/s Acousticness Correlation")
```

```python
#the regression line has negative slope
#inversely proportional
# weak correlation
```

Creating new column Year from release date

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years= df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete= True, aspect=2, height= 5, kind="hist" ).set(title="Number of songs per year")
```

Barplot for duration of songs

```python
total_dr = df_tracks.duration
total_dr
```

```python
fig_dims = (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth= False).set(title="Years v/s Duration")
plt.xticks(rotation=90)
```

```python
#duration of songs is decreasing
```

Line Plot to analyse average duration of songs over the years

```python
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
figs_dim=(10,6)
fig, ax=plt.subplots(figsize=figs_dim)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Years v/s Duration")
#plt.xticks(rotation=60)
```
